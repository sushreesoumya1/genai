# Let us Experience Feature Enginneering
---
Extract Deep Information from Existing Data by Populating New Columns from Existing Ones.

```python
nba.head()
```

```python
nba.Salary.head()
```

```python
# Convert Salary into Millions (To Improve the Readability )

nba.Salary / 1000000         # Broadcasting
```

```python
def convertToMill( sal ):

  return sal / 1000000
```

```python
nba.Salary.apply( convertToMill  )

# what does the above expression means ?
# answer: apply function calls convertToMill function for each value of Salary Column.
# apply function passes each value of Salary column into variable Sal of convert to Mill function.
```

```python
def convertToMill( sal ):

  return f'{sal / 1000000} Mill.'
```

```python
# Below Expression Adds a New Column  (Salary(mill.)) and stores the converted values.

nba['salary(mill)']= nba.Salary.apply( convertToMill  )
```

```python
nba.head(3)
```

```python
del nba['Salary']
```

task :  using the  `weights(kgs)` column you are asked to generate a new Column `playerWeightCategory`

and generate the values as per the Standards Below:

if weight >120                'overweight'
if weight between 90 and 120 'normalweight'
if weight < 90               'underweight'


```python
# you are asked to reply with 'how many players belong to each category'
```

```python
def weightCategoriser( wght ):

  if wght > 120:
    return 'overweight'

  elif wght >=90:
    return 'normalweight'

  else:
    return 'underweight'
```

```python
nba['weightCategory'] = nba['Weight(kgs)'].apply( weightCategoriser)
```

# HeatMap

heatmap is used to visualize 2D Data.

```python
flights_df= sns.load_dataset('flights')
```

# i will be using .pivot() here to better restructure the data to improve Readability and Analysis both.
